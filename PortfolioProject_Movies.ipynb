```python
# Import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
```

```python
# Read in data
# https://www.kaggle.com/danielgrijalvas/movies

df = pd.read_csv(r'C:\Users\Blue\Documents\PortfolioProjects\movies\movies.csv')
```

```python
#Check for missing data
df.isnull().sum()
```

```python
# check column types
df.dtypes
```

```python
# change data type of budget and gross columns from float to int

df['budget'] = np.nan_to_num(df['budget']).astype('int64')
df['gross'] = np.nan_to_num(df['gross']).astype('int64')
```

```python
df.head()
```

```python
#create correct year column
df['yearcorrect'] = df['released'].astype(str).str[-4:]
```

```python
df.head()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# edit display window to show all values in the data set
pd.set_option('display.min_rows', None)
```

```python
#check for duplicates

df.drop_duplicates()
```

```python
# Reset window size to something manageable because no one wants to waste time loading up thousands of rows
pd.set_option('display.min_rows', 20)
```

```python
# Build a scatterplot to observe correlation between budget and gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
df.head()
```

```python
# Plot 
sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# Test for correlation
df.corr(method='pearson')
```

```python
# High correlation between budget and gross
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Look at company. Does it have a correlation
df.head()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
